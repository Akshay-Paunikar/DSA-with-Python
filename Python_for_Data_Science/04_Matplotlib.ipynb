### Matplotlib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+. general

Matplotlib is a plotting library used for creating static, interactive, and animated visualizations in Python. It's highly customizable and provides control over the creation of various types of plots.

Some of the major Pros of Matplotlib are:

 -  Generally easy to get started for simple plots
 -  Support for custom labels and texts
 -  Great control of every element in a figure
 -  High-quality output in many formats
 -  Very customizable in general

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# simple examples
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# scatter plot
plt.figure(figsize=(6,4))
plt.scatter(x, y, c='g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("scatter plot")
plt.show()
```

```python
# exponential
y = x**2
```

```python
# let's plot the data using plt.plot
plt.figure(figsize=(7,4))
plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("scatter plot")
plt.show()
```

```python
# create subplots
# first scatter plot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("scatter plot")

# second scatter plot
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("scatter plot")

# third scatter plot
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("scatter plot")

# fourth scatter plot
plt.subplot(2,2,4)
plt.plot(x,y,'go')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("scatter plot")

plt.tight_layout()
plt.show()
```

```python
# using an equation
x = np.arange(1,11) 
y = 3 * x + 5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure.
plt.tight_layout()
plt.show()
```

```python
# Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
# Histogram
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```python
# Boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()
```

```python
# Pie Chart
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [315, 180, 295, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
